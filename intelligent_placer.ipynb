# Отчет. Демонстрация работы алгоритм

```python
import logging
from colorama import Fore
from intelligent_placer_lib.intelligent_placer import check_image
from dataset.dataset import get_true_images, get_false_images
# Custom threshold for placer
from skimage.filters.thresholding import threshold_yen

# Log for report
logging.basicConfig(level=logging.INFO)
```

## Точность на позитивных данных

Проверяется точность работы плейсера на данных, размеченных, как True (предметы можно поместить в многоугольник)

```python
positive_images = get_true_images()
true_positive = 0
for image in positive_images:
    print(Fore.GREEN + image)
    if check_image(image, with_images=True, threshold_delimeter=1.5, threshold=threshold_yen):
        true_positive += 1

print(f"True accuracy: {true_positive / len(positive_images)}")
```

True accuracy = 0.82. В целом, неплохой результат, ошибки связаны с тем, что я аппроксимирую предметы сложной формы bounding box-ми, не учитываю их форму, соответственно компоновка вплотную невозможна. Также можно заметить, что предметы выделяются не совсем точно, из-за дыр и пробелов в масках, но основной массив предмета выделяется хорошо, поэтому это не влияет на результат.

## Точность на негативных данных

Проверяется точность работы плейсера на данных, размеченных, как False (предметы нельзя поместить в многоугольник)

```python
negative_images = get_false_images()
true_negative = 0
for image in negative_images:
    print(Fore.RED + image)
    if not check_image(image, with_images=True, threshold_delimeter=1):
        true_negative += 1

print(f"False accuracy: {true_negative / len(negative_images)}")
```

Как видим, все изображения, размеченные как False были правильно интерперетированы.

## Общая точность

```python
print(f"accuracy: {(true_negative + true_positive) / (len(positive_images) + len(negative_images))} ")
```
